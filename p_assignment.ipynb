```python
# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
```

```python
# Ans = In Pandas, a DataFrame is a two-dimensional, size-mutable, and heterogeneous tabular data structure.
# It can be thought of as a spreadsheet or SQL table, where data is organized into rows and columns. Each column in a DataFrame
# can have a different data type (e.g., integers, floats, strings).

# On the other hand, a Series in Pandas is a one-dimensional labeled array that can hold data of any data type, similar to a 
# single column in a DataFrame.
```

```python
# Example :-
# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
```

```python
# Q7. Create a DataFrame using multiple Series. Explain with an example.
```

```python
# Ans => Certainly! In Python, you can create a DataFrame using multiple Series by using a library called Pandas. Pandas is a popular data manipulation library
# that provides data structures like DataFrames for working with tabular data.
# Here's an example of how to create a DataFrame using multiple Series:-

# Creating two Series
series1 = pd.Series([1, 2, 3, 4])
series2 = pd.Series(['A', 'B', 'C', 'D'])
```

```python
# Creating a DataFrame by combining the Series
df = pd.DataFrame({'Column1': series1, 'Column2': series2})

# Displaying the DataFrame
print(df)
```
